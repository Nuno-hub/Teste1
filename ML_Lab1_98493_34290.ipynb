#### 2.1.2

```python
import numpy as np

def polynomial_fit(x,y,P):
    A = np.zeros((len(x),P+1));  # Initializes the A matrix, corresponding to the matrix form of the normal equations
    
    for i in range(P+1):
        for k in range(len(x)):  # Brute-force nested for loops to fill the matrix
            A[k,i] = x[k]**i;
    
    At = np.transpose(A);
    B = np.dot(At,y);
    Coef = np.dot(np.linalg.inv(np.dot(At,A)),B);  # The coefficients are computed by solving the system
    
    return np.round(Coef,3);  # Rounds the coefficients
```

#### 2.1.3

```python
x = np.load('data1_x.npy');  #Import the data and showcase it
y = np.load('data1_y.npy');

print(x);
print(y);
```

```python
x = x[:,0];  # Convert the data to a proper format
print(x);
```

```python
y = y[:,0];
print(y);
```

```python
c = polynomial_fit(x,y,1);# Apply the polynomial fit and print the coefficients. Left: B0
print(c);
```

```python
def SSE_linear(x,y,coef):
    prediction = np.zeros(len(y));    # This function computes the SSE, receiving the data arrays and the previously obtained
                                       # coefficients as arguments    
    for i in range(len(y)):
        prediction[i] = coef[1]*x[i] + coef[0];
    
    SSE = np.sum((y-prediction)**2);
    
    print("Coefficients: B0 = ",coef[0]," B1 = ",coef[1]);
    print("SSE = ", np.round(SSE,3));
    
    return prediction;
```

```python
res = SSE_linear(x,y,c);
```

```python
import matplotlib.pyplot as plt

plt.scatter(x,y,label='Data'); # Plot both the data and the fit on the same graph
plt.plot(x,res,'r',label='Fit');
plt.legend();
```

It can be concluded that the model provides a good fit for the data. The SSE is very low, and the errors seem to be evenly distributed

```python
def SSE_quadratic(x,y,coef):
    prediction = np.zeros(len(y));  # Computes the SSE for a second-order fit
    
    for i in range(len(y)):
        prediction[i] = coef[2]*x[i]**2 + coef[1]*x[i] + coef[0];
    
    SSE = np.sum((y-prediction)**2);
    
    print("Coefficients: B0 = ",coef[0]," B1 = ",coef[1]," B2 = ",coef[2]);
    print("SSE = ", np.round(SSE,3));
    
    return prediction;
```

```python
res2 = SSE_quadratic(x2,y2,c2);
```

```python
plt.scatter(x2,y2,label='Data');  # Plot the data and the fit
plt.scatter(x2,res2,label='Fit');
plt.legend();
plt.title('Second order polynomial fit');
```

Based on the plot and the SSE obtained, it can be concluded that a second order polynomial is an adequate model to describe the data. There are a few points right in the middle of the plot that don't seem to fit very well, but this can be attributed to noise, since we do know the data was noisy. Otherwise, the model performs very well, and the SSE is relatively small.

##### 2.2.2

Ridge regression uses a regularization term that is the squared Euclidean norm of coeficients, multiplied by a constant $\lambda$,

\begin{equation}
\begin{aligned}
\\
\hat{\beta}_{ridge} = \underset{\beta}{\text{arg min}} \|y-X\beta\|^2 + \lambda\|\beta\|^2
\end{aligned}
\end{equation}

The term $ \|\beta\|^2 $ penalizes the use of large coefficients and is denoted as a regularization term. This criterion aims to represent the data, keeping the coefficients small.
$\lambda$ represents the trade of between both objectives, the data fit and the regularization.

Lasso regression uses a regularization term that is the $l_{1}$ norm of coefficients, multiplied by a constant,

\begin{equation}
\begin{aligned}
\\
\hat{\beta}_{lasso} = \underset{\beta}{\text{arg min}} \|y-X\beta\|^2_2 + \lambda\|\beta\|_1
\end{aligned}
\end{equation}

The regularization term is a good aproximation of the $\lambda$ "$l_{0}$ norm" where "$l_{0}$ norm" represents the number of non-zero coeficients and is dificult to solve numerically.

This equation cannot be solved analytically, and we must resort to convex optimization methods to numerically solve this problem. When a feature is not important, the corresponding coefficient is made equal to zero and often get a sparse vector $ \beta $. 
This procedure can be understood as a feature selection as the features with coefficient equal to zero will not be taking into account.
    

#### 2.2.3 & 2.2.4

```python
import sklearn.linear_model as skl
```

```python
ridge_coef = [];  # Lists for the Ridge and Lasso coeficients
lasso_coef = [];
alpha = np.arange(0.001,10,0.001);

for i in alpha:  # Loop to fit models to the data in the specified range of alpha values
    ridge = skl.Ridge(alpha=i, max_iter=10000, fit_intercept=False);
    lasso = skl.Lasso(alpha=i, max_iter=10000, fit_intercept=False);
    ridge.fit(x4,y4);
    lasso.fit(x4,y4);
    ridge_coef.append(np.array(ridge.coef_));  # Appending the coefficients to the previously created lists
    lasso_coef.append(np.array(lasso.coef_));
```

#### 2.2.7

Alpha = 0.5 was the value that was initially chosen. However, it proved to be too high and produced poor results, with the SSE being much larger than the one for the LS fit. After a more careful observation of the plots, alpha = 0.1 was chosen, and this time it yielded good results, as shown and explained below.

```python
lasso2 = skl.Lasso(alpha=0.1,max_iter=10000,fit_intercept=False);   # Lasso model with alpha = 0.1
lasso2.fit(x4,y4);
print(lasso2.coef_);
```

```python
def Lasso_plot(x,y,lasso_0,lasso_05):   # Function to calculate the predicted results using the coefficients obtained from the
    prediction1 = np.zeros(len(y));    # Lasso and LS methods. lasso_0 is the list with the coefficients for alpha = 0.
    prediction2 = np.zeros(len(y));   # lasso05 is the list containing the coefficients for alpha = 0.1
    
    for i in range(len(y)):
        prediction1[i] = lasso_0[0]*x[i,0]+lasso_0[1]*x[i,1]+lasso_0[2]*x[i,2];
        prediction2[i] = lasso_05[0]*x[i,0]+lasso_05[1]*x[i,1]+lasso_05[2]*x[i,2];
    
    SSE1 = np.sum((y-prediction1)**2);  # The function also calculates the SSE for both models
    SSE2 = np.sum((y-prediction2)**2);
    
    print("SSE with alpha = 0:  ", np.round(SSE1,2));
    print("SSE with alpha = 0.1: ", np.round(SSE2,2));
    
    return prediction1, prediction2;
```

```python
p1, p2 = Lasso_plot(x4,y4,lasso.coef_,lasso2.coef_)
```

```python
plt.plot(y4,'b*',label='Data');   # Data plotting
plt.plot(p1,'r*',label='LS fit');
plt.plot(p2,'g*',label='Lasso fit');
plt.legend();
plt.title('Data and fits');
```

We can conclude that the choice of the adequate value for parameter Alpha is an incredibly important one, since the performance of the Lasso method stems directly from it. The purpose of this method is to identify irrelevant features, thus be able to produce a model that isn't affected by them (the coefficient will be null). However, by choosing a larger alpha, one puts less emphasis on the least-squares term. Therefore, the fit won't be as good, increasing the SSE. In essence, what we want is to choose a value of alpha that forces the coefficient affected to the irrelevant feature to be zero, but doesn't increase the SSE too much. The chosen value 0.1 proves to be a good compromise between these two goals.
